```
# Step 2: Upload project files
import os
os.makedirs('/content/nokia_storage/fileprovider_src/res/xml', exist_ok=True)
os.chdir('/content/nokia_storage')

from google.colab import files
print('Upload these files: main.py, database.py, buildozer.spec')
print('(Select all 3 at once)')
uploaded = files.upload()
```

```
# Step 3: Create FileProvider XML (needed for camera/sharing)
filepaths_xml = '''<?xml version="1.0" encoding="utf-8"?>
<paths>
    <external-path name="external_files" path="." />
    <files-path name="internal_files" path="." />
    <cache-path name="cache_files" path="." />
    <external-files-path name="external_app_files" path="." />
</paths>'''

with open('fileprovider_src/res/xml/filepaths.xml', 'w') as f:
    f.write(filepaths_xml)

print('Project ready!')
!ls -la
```

```
# Step 4: Build APK (~12-15 min first time)
!buildozer android debug 2>&1 | tail -30
```

```
# Step 5: Download APK
import glob
apks = glob.glob('bin/*.apk')
if apks:
    print(f'APK built: {apks[0]}')
    files.download(apks[0])
else:
    print('Build failed. Check output above for errors.')
```
